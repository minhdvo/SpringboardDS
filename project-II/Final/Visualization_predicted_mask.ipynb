```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

## Validation set
- Validation set has 1200 seismic images file in training set
- ids_valid.npy file consisted all id file of validation set
- preds_valid.npy file included all information of mask images for validation set

```python
ids_valid = np.load('ids_valid.npy')
ids_valid
```

```python
preds_valid= np.load('preds_valid.npy')
preds_valid[1]
```

```python
preds_valid.shape
```

```python
i=110
print('File ID: ' + ids_valid[i])
plt.figure(figsize=(10, 10))

img = mpimg.imread('train/images/'+ids_valid[i]+'.png')
plt.subplot(131)
plt.title('Seismic image')
plt.grid(False)
plt.imshow(img)

plt.subplot(132)
img = mpimg.imread('train/masks/'+ids_valid[i]+'.png')
plt.title('Truth mask')
plt.grid(False)
plt.imshow(img,cmap='gray')

plt.subplot(133)
plt.imshow(preds_valid[i],cmap='gray')
plt.title('Predicted mask')
plt.grid(False)
plt.show()
```

- Test set has 18000 seismic images file

```python
preds_test.shape
```
