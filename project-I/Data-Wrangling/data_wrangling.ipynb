# 2. Total_Renewable_Energy_Consumption_Annual

```python
# Total_Renewable_Energy_Consumption_Annual
df1 = pd.read_csv('renew/1-Total_Renewable_Energy_Consumption_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df1.columns = ['total-renewable-consumption']
df1 = df1.sort_index()

# 2.1 Total_Renewable_Energy_Production_Annual
df2 = pd.read_csv('renew/2-Total_Renewable_Energy_Production_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df2.columns = ['total-renewable-production']
df2 = df2.sort_index()

# 2.2. Conventional_Hydroelectric_Power_Consumed_by_the_Electric_Power_Sector_Annual
df3 = pd.read_csv('renew/3-Conventional_Hydroelectric_Power_Consumed_by_the_Electric_Power_Sector_Annual.csv',
                  index_col='Year',skiprows=4,parse_dates=True)
df3.columns = ['Hydroelectric-power']
df3 = df3.sort_index()

# 2.3. Hydroelectric_Power_Consumption_per_Production_Annual
df4 = pd.read_csv('renew/4-Hydroelectric_Power_Consumption_per_Production_Annual.csv',index_col='Year',
                  skiprows=4,parse_dates=True)
df4.columns = ['Hydroelectric-power-Con-Pro']
df4 = df4.sort_index()

# 2.4. Geothermal_Energy_Consumption_per_Production_Annual
df5 = pd.read_csv('renew/5-Geothermal_Energy_Consumption_per_Production_Annual.csv',index_col='Year',
                  skiprows=4,parse_dates=True)
df5.columns = ['Geothermal_Energy_Consumption_per_Production']
df5 = df5.sort_index()
df5 = df5.fillna(0)

# 2.5. Geothermal_Energy_Consumed_by_the_Electric_Power_Sector_Annual
df6 = pd.read_csv('renew/6-Geothermal_Energy_Consumed_by_the_Electric_Power_Sector_Annual.csv',
                  index_col='Year',skiprows=4,parse_dates=True)
df6.columns = ['Geothermal_Energy_Consumption_by_the_Electric_Power']
df6 = df6.sort_index()
df6 = df6.fillna(0)

# 2.6. Solar_per_PV_Energy_Consumption_per_Production_Annual
df7 = pd.read_csv('renew/7-Solar_per_PV_Energy_Consumption_per_Production_Annual.csv',index_col='Year',
                  skiprows=4,parse_dates=True)
df7.columns = ['Solar_per_PV_Energy_Consumption_per_Production']
df7 = df7.sort_index()
df7 = df7.fillna(0)

# 2.7. Total_Biomass_Energy_Consumed_by_the_Electric_Power_Sector_Annual
df8 = pd.read_csv('renew/8-Total_Biomass_Energy_Consumed_by_the_Electric_Power_Sector_Annual.csv',
                  index_col='Year',skiprows=4,parse_dates=True)
df8.columns = ['Total_Biomass_Energy_Consumed_by_the_Electric_Power']
df8 = df8.sort_index()

# 2.8. Total_Biomass_Energy_Production_Annual
df9 = pd.read_csv('renew/9-Total_Biomass_Energy_Production_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df9.columns = ['Total_Biomass_Energy_Production']
df9 = df9.sort_index()

# 2.9. Total_Renewable_Energy_Consumed_by_the_Electric_Power_Sector_Annual
df10 = pd.read_csv('renew/10-Total_Renewable_Energy_Consumed_by_the_Electric_Power_Sector_Annual.csv',
                   index_col='Year',skiprows=4,parse_dates=True)
df10.columns = ['Total_Renewable_Energy_Consumed_by_the_Electric_Power']
df10 = df10.sort_index()

# 2.10. Waste_Consumption_for_Electricity_Generation_and_Useful_Thermal_Output_Electric_Power_Sector_Annual
df11 = pd.read_csv('renew/11-Waste_Consumption_for_Electricity_Generation_and_Useful_Thermal_Output_Electric_Power_Sector_Annual.csv',
                   index_col='Year',skiprows=4,parse_dates=True)
df11.columns = ['Waste_Consumption_for_Electricity']
df11 = df11.sort_index()
df11 = df11.fillna(0)

# 2.11. Wind_Energy_Consumed_by_the_Electric_Power_Sector_Annual
df12 = pd.read_csv('renew/12-Wind_Energy_Consumed_by_the_Electric_Power_Sector_Annual.csv',
                   index_col='Year',skiprows=4,parse_dates=True)
df12.columns = ['Wind_Energy_Consumed_by_the_Electric_Power']
df12 = df12.sort_index()
df12 = df12.fillna(0)

# 2.12. Wind_Energy_Consumption_per_Production_Annual
df13 = pd.read_csv('renew/13-Wind_Energy_Consumption_per_Production_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df13.columns = ['Wind_Energy_Consumption_per_Production']
df13 = df13.sort_index()
df13 = df13.fillna(0)

# 2.13. Wood_Consumption_for_Electricity_Generation_and_Useful_Thermal_Output_Electric_Power_Sector_Annual
df14 = pd.read_csv('renew/14-Wood_Consumption_for_Electricity_Generation_and_Useful_Thermal_Output_Electric_Power_Sector_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df14.columns = ['Wood_Consumption_for_Electricity']
df14 = df14.sort_index()

# 2.14. Biomass_Exports_Annual
df15 = pd.read_csv('renew/15-Biomass_Exports_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df15.columns = ['Biomass_Exports']
df15 = df15.sort_index()
df15 = df15.fillna(0)


# 2.15. Biomass_Imports_Annual
df16 = pd.read_csv('renew/16-Biomass_Imports_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df16.columns = ['Biomass_Imports']
df16 = df16.sort_index()
df16 = df16.fillna(0)

# 2.16. Biofuels_Consumption_Annual
df17 = pd.read_csv('renew/17-Biofuels_Consumption_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df17.columns = ['Biofuels_Consumption']
df17 = df17.sort_index()
df17 = df17.fillna(0)

# 2.17. Biofuels_Production_Annual
df18 = pd.read_csv('renew/18-Biofuels_Production_Annual.csv',index_col='Year',skiprows=4,parse_dates=True)
df18.columns = ['Biofuels_Production']
df18 = df18.sort_index()
df18 = df18.fillna(0)
```
