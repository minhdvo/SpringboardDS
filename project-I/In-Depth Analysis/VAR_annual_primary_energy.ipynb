### This notebook presented the prediction of annual primary energy consumption by VAR model

```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Load data

```python
df = pd.read_csv('data_3.csv',index_col=[0],parse_dates=True)
df.info()
```

## Important features

```python
data = df[['primary-energy','CO2','Electricity-price','crude_oil']].dropna()
```

```python
# Perform Dickey-fuller test  for stationality
from statsmodels.tsa.stattools import adfuller
```

```python
for i in data.columns: 
    x = data[i].values
    result = adfuller(x)
    print('\033[1m' + i + '\033[0m')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('')
```

## Transform data

```python
diff = data.diff().dropna()
diff.head()
```

```python
for i in diff.columns: 
    x = diff[i].values
    result = adfuller(x)
    print('\033[1m' + i + '\033[0m')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('')
```

```python
diff['CO2']=diff['CO2'].diff()
diff.head()
```

```python
result = adfuller(diff['CO2'].dropna())
print('CO2')
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
diff = diff.dropna()
diff.info()
```

```python
diff.plot(subplots=True)
```

## Split data

```python
x_train = diff.iloc[:58,:]
y_train = diff.iloc[:58,0]

x_test = diff.iloc[58:,:]
y_test = diff.iloc[58:,0]
```

```python
x_train.shape
```

```python
x_test.shape
```

## Build Model

```python
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
```

```python
# make a VAR model
model = VAR(x_train,freq='AS-JAN')
```

```python
model.select_order(3).summary()
```

```python
res = model.fit(maxlags=3, ic='aic')
```

```python
res.summary()
```

```python
lag_order = res.k_ar
lag_order
```

## Predict and Compare with test set

```python
pred = res.forecast(x_train.values[-lag_order:], 9)
pred
```

```python
res.plot_acorr()
plt.show()
```

```python
res.plot_forecast(9)
plt.show()
```

```python
y_pre = pd.DataFrame(pred).iloc[:,0]
y_pre
```

```python
plt.plot(y_test.index,y_pre.values)
y_test.plot()
plt.legend(['fit','test'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
math.sqrt(mean_squared_error(y_pre.values, y_test.values))
```
